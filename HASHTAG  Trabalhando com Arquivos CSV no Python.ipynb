```python
# Usaremos o arquivo csv  adverising.csv
# Usando PANDAS

import pandas as pd

caminho_arquivo = r'C:\Users\User\Desktop\RPA\advertising.csv'

advert = pd.read_csv(caminho_arquivo)

display(advert)
```

```python
# ANALISAR COMO TEXTO COM O METODO INTERNO DO PYTHON

import csv
# Importante usar a estrutura With para que o arquivo seja aberto e FECHADO após utilização
with open(caminho_arquivo,"r") as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter=",")
    
    ## Percorrer o arquivo
    #for linha in arquivo_csv:
    #    print(linha)
    
    for i, linha in enumerate(arquivo_csv):
        if i == 0:
            print("CABEÇALHO: ",linha)
        else:
            print(i,linha)
```
