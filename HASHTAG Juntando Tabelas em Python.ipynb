```python
# Usando PANDAS
import pandas as pd

# Colocando os Caminhos das tabelas em LOCAL
caminho = r'C:\Users\User\Desktop\RPA'
arquivo_py = caminho+'\python.xlsx'
arquivo_cie = caminho+'\cienciadados.xlsx'
```

```python
# Visualizando a Python.xlsx
tabela_py = pd.read_excel(arquivo_py)
display(tabela_py)
```

```python
# Visualizando a cienciadados.xlsx
tabela_cie = pd.read_excel(arquivo_cie)
display(tabela_cie)
```

<h2> Concat

```python
#####Adicionar uma tabela embaixo da outra
### Precisa ser passado como lista!
lista_tabelas = [tabela_py,tabela_cie]

tabela_cursos = pd.concat(lista_tabelas)
### Alternativamente:
# tabela_cursos = pd.concat([tabela_py,tabela_cie])
display(tabela_cursos)
```

```python
## Repare que os Indíces se mantiveram das tabelas originais
## Como os indices nao eram informações relevantes, iremos resetar os indices

# Usar o parametro:   ignore_index

tabela_cursos = pd.concat(lista_tabelas, ignore_index = True)
display(tabela_cursos)
```

```python
## Depois de feito a concatenação, caso você queira saber de onde vem a info, usaremos o parametro KEYS 
## para cada tabela na lista

## Por esse metodo criar um indice, não poderá ser usado o IGNORE_INDEX
tabela_cursos = pd.concat(lista_tabelas, keys=["python","cienciadados"])
display(tabela_cursos)
```

<h2>MERGE</h2>

Mescla tabelas

```python
tabela_produtos = pd.merge(tabela_py, tabela_cie)
display(tabela_produtos)

# A Tabela resultante veio vazia pois o python não conseguiu identificar uma conexão válida sem interferencia
```

```python
## Quais clientes compraram os 2 produtos

# Para corrigir o erro acima vamos falar pro Python que ele deve pegar a Coluna CLIENTES ( on )
tabela_produtos = pd.merge(tabela_py, tabela_cie, on='Cliente')
display(tabela_produtos)

#Note que foram acrescentados Colunas mostrando o que tem em uma tabela (x) e o que tem na outra (y)
```

```python
## E Se os nomes das colunas forem diferentes nas tabelas?
# Aí precisariamos determinar quais são as duas colunas com os parametros LEFT_ON e RIGHT_ON
tabela_produtos = pd.merge(tabela_py, tabela_cie, left_on='Cliente',right_on='Cliente')
display(tabela_produtos)
```

```python
## PARAMETRO HOW
# Por padrão, o merge usa o paremetro INNER (pega a interseção de dados, dados que contem nos dois conjuntos)


# LEFT
# Mescla os dois resultados, mas mantem as informações da Tabela da Esquerda

tabela_produtos = pd.merge(tabela_py, tabela_cie, on='Cliente',how='left')
display(tabela_produtos)
```

```python
# RIGHT
# Mescla os dois resultados, mas mantem as informações da Tabela da Direita

tabela_produtos = pd.merge(tabela_py, tabela_cie, on='Cliente',how='right')
display(tabela_produtos)
```

```python
# outer
# Mescla os dois resultados, mas mantem todas as informações

tabela_produtos = pd.merge(tabela_py, tabela_cie, on='Cliente',how='outer')
display(tabela_produtos)
```
